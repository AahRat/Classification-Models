```python
from sklearn.model_selection import train_test_split
```

Train Test Split --> Allows to choose test and train ratio

```python
x = iris.iloc[:,0:3]
```

```python
x
```

```python
x = x.values
```

```python
y = iris.iloc[:,4]
```

```python
y
```

```python
y = y.values
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y, test_size=0.3)
```

Test_Size = 0.3 --> Test Data=30% and Trained Data = 70%

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

Importing Model

```python
treeModel = DecisionTreeClassifier()
```

```python
treeModel.fit(x_train,y_train)
```

```python
y_predict = treeModel.predict(x_test)
```

Generating Prediction Values using the testing values

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_predict, y_test)
```

```python
for i in range(0,4):
  x = iris.iloc[:,0:i+1]
  x = x.values
  y = iris.iloc[:,4]
  x_train,x_test, y_train,y_test = train_test_split(x,y, test_size=0.3)
  treeModel = DecisionTreeClassifier()
  treeModel.fit(x_train,y_train)
  y_predict = treeModel.predict(x_test)
  acc = accuracy_score(y_predict, y_test)
  print(f"For {i} column's the acucracy score is: {acc}")
```
